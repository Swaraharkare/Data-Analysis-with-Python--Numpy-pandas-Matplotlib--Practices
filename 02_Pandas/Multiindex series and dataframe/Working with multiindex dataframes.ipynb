```python
# how to create multiindex object
# 1. pd.MultiIndex.from_tuples()
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
```

```python
branch_df3 = pd.DataFrame(
    [
        [1,2,0,0],
        [3,4,0,0],
        [5,6,0,0],
        [7,8,0,0],
        [9,10,0,0],
        [11,12,0,0],
        [13,14,0,0],
        [15,16,0,0],
    ],
    index = multiindex,
    columns = pd.MultiIndex.from_product([['delhi','mumbai'],['avg_package','students']])
)

branch_df3
```

```python
branch_df3.iloc[:,1:3]  #branch_df3.iloc[all rows,1 col to col 3]
```

```python
# Extracting both
branch_df3.iloc[[0,4],[1,2]]   #branch_df3.iloc[[rows index no],[col index no]]
```

```python
# sort index
# both -> descending -> diff order
# based on one level
branch_df3.sort_index(ascending=False)
```

```python
branch_df3.sort_index(ascending=[False,True])
```

```python
branch_df3.sort_index(level=1,ascending=[False])  #sorting on level 1 
```

```python
# multiindex dataframe(col) -> transpose-----row to col --col to row
branch_df3.transpose()
```

```python
# swaplevel----swaping level
branch_df3.swaplevel()
```

```python
# swaplevel
branch_df3.swaplevel(axis=1)
```
