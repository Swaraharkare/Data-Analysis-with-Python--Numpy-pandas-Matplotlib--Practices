```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('movies_dataframe.csv')
movies
```

```python
#iloc-searches using index position
#single row---series
movies.iloc[0]
```

```python
#multiple rows---dataframe
movies.iloc[0:5]
```

```python
#fancy indexing
movies.iloc[[0,4,7]]
```

```python
#loc-searching using index labels
#single row
movies.loc[2]
```

```python
#multiple rows
movies.loc[2:6]
```

```python
#fancy indexing
movies.loc[[2,6,1]]
```

```python
####Selecting both rows and cols
movies.iloc[0:3,0:3]
```

```python
movies.loc[0:2,'title_x':'poster_path']
```
