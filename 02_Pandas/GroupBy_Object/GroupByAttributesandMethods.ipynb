```python
# GroupBy Attributes and Methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies
```

```python
genres=movies.groupby('Genre')
```

```python
# find total number of groups -> len
len(movies.groupby('Genre'))  #movies['Genre'].nunique()
```

```python
# find items in each group -> size
movies.groupby('Genre').size()   #movies['Genre'].value_counts()
```

```python
# first()/last() -> nth item --- of each group
genres = movies.groupby('Genre')
genres.first()
```

```python
genres.last()
```

```python
genres.nth(6)
```

```python
# get_group -> vs filtering
genres.get_group('Fantasy')
#movies[movies['Genre'] == 'Fantasy']
```

```python
# groups-----group index number of each genre
genres.groups
```

```python
# describe---mathematical summerize
genres.describe()
```

```python
# sample
genres.sample() #fetch one random movie from each genre
genres.sample(2,replace=True) 
#fetch two random movies from each genre---replace=True if there is one movie in any genre then we can fetch same movie two times 
```

```python
# nunique
genres.nunique()
```

```python
# agg method----apply different aggregation function on different column
# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
#genres.sum()
```

```python
# passing list....apply all below aggre fun on each column 
genres.agg(['min','max','sum'])
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
df = pd.DataFrame(columns=movies.columns)
for group, data in genres:
    # Find the row(s) with the highest rating in the current genre group
    max_row = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    
    # Concatenate the new row(s) with the existing empty/growing DataFrame
    df = pd.concat([df, max_row], ignore_index=True)

df


# data is dataframe of each genre
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
# find number of movies starting with A for each group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
  
```

```python
genres.apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

```python
# groupby on multiple cols
duo = movies.groupby(['Director','Star1'])
duo
# size
duo.size()
# get_group
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find the most earning actor->director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# agg on multiple groupby
duo.agg(['min','max','mean'])
```
