```python
#The NumPy transpose operations swap rows and columns in 2D arrays or rearrange axes in higher-dimensional arrays 
```

```python
import numpy as np

a=np.arange(12).reshape(3,4)
print(a)
```

```python
#temperory operations
#transpose()----convert col into rows and rows into columns

a.transpose()
```

```python
#3D array
b = np.arange(1,13).reshape(2,2,3)
b
```

```python
#transpose() 3D array
a = np.arange(24).reshape(2, 3, 4)
print(a)
a.transpose()
```
