```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("sample2.txt", index_col='idx')
# Convert to Series
series = cars.stack()
print(type(series))
print(series)
```

```python
# what if an index does not exist
series[11]= 'Tata Nexon EV'
```

```python
print(series)
```

```python
#using slicing

series[2:4]=['Tata   Punch','Maruti Brezza']
print(series)
```

```python
#Fancy indexing
series.iloc[[5,6]]=['Mahindra Scorpio N','Toyota Fortuner']
print(series)
```

```python
#using index label
series[1]=['Audi A4']
print(series)
```
