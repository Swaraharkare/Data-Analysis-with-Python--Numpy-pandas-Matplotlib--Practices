```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Colored Scatterplots
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
#c means colour...cmap is colour theme...alpha opacity (darkness of points)
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.9)
#plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer')
#plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```
