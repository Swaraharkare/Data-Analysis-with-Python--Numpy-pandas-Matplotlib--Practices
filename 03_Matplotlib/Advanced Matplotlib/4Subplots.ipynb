```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batters = pd.read_csv('batter.csv')
```

```python
# A diff way to plot graphs
batters.head()
```

```python
plt.scatter(batters['avg'],batters['strike_rate'])
```

```python
#Another way to plot scatter plot
fig, ax=plt.subplots()
ax.scatter(batters['avg'],batters['strike_rate'])
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
plt.subplots(nrows=3,ncols=2)
```

```python
# batter dataset
#first way to plot graphs

fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
#second way to plot subplots
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
#third way to plot graphs
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)   # 2 rows 2 cols 1st graph
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)   # 2 rows 2 cols 2nd graph
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])  # 2 rows 2 cols 3rd graph
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,1,1)   # 3 rows 1 cols 1st graph
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(3,1,2)   # 3 rows 1 cols 2nd graph
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(3,1,3)
ax3.hist(batters['avg'])  # 3 rows 1 cols 3rd graph
```
