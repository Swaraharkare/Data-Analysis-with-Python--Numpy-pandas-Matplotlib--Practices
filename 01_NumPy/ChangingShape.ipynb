```python
import numpy as np
a = np.arange(8)
print(a)
```

```python
#reshape()
a.reshape(4,2)
```

```python
#flat
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(array_2d)
for item in array_2d.flat:
    print(item)
```

```python
#ravel()
print(np.ravel(array_2d))
```

```python
#pad()
array = np.array([1, 2, 3])
padded_array = np.pad(array, pad_width=2, mode='constant')
print(padded_array)
```
