```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("sample.txt",index_col='Car_Model')
# Convert to Series
series = cars.stack()
print(type(series))
print(series)
```

```python
#head----head() give first five rows bydefault---
series.head()
```

```python
series.head(8)
```

```python
#tail---give last five rows bydefault
series.tail()
```

```python
#gives last 8 rows
series.tail(8)
```

```python
#sample--- sample() fetch one row randomly
print(series.sample())
print()
#sample() fetch five row randomly
print(series.sample(5))
```

```python
#values_counts()
series.value_counts()
```

```python
type(series)
```

```python
#sort_values()---bydefault ascending--by values
series.sort_values()

#descending
series.sort_values(ascending=False)
```

```python
#sort_values() by values and head()-----to give largest Price_USD from  dataset-----not permantly sort
series.sort_values(ascending=False).head(1)

#to take only value
series.sort_values(ascending=False).head(1).values[0]
print(series.sort_values(ascending=False).head(1).values[0])
                                                      
#above is method chaning
```

```python
#to do permanently sorting of dataset by values
series.sort_values(ascending=False,inplace=True)
```

```python
#sort_index()---sort index in ascending order by index
series.sort_index()
```

```python
#sort_index()---sort index in descending order by index--inplace() to sort_index permantaly
series.sort_index(ascending=False,inplace=True)
series
```
