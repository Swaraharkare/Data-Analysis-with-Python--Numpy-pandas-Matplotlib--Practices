```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sample.txt",index_col='Car_Model')
# Convert to Series
series = df['Price_USD']
print(type(series))
print(series)
```

```python
#len/type/dir/max/min/sorted
print(len(series))
print(type(series))
print(max(series))
print(min(series))
print(sorted(series))
print(dir(series)) #shows all functions of series
```

```python
#type conversion
print(list(series))
print(dict(series))
```

```python
#membership operator----working on index
'Chevrolet Corvette' in series

#working on values
23950 in series.values
```

```python
#looping---printing values
for i in series:
    print(i)
```

```python
#printing index
for i in series.index:
    print(i)
```

```python
#aithmetic operators(broadcasting)
series+10000
```

```python
#relational operator
series >=30000
```
