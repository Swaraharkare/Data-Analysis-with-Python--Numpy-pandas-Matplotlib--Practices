```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batters = pd.read_csv('batter.csv')
```

```python
fig = plt.figure()

ax = plt.subplot(projection='3d')
```

```python
#3D Scatter Plots

batters

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],marker='+')
ax.set_title('IPL batsman analysis')

ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('SR')
```

```python
#3D Line Plot

x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
```

```python
#3D Surface Plots

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
