```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("sample.txt",index_col='Car_Model')
# Convert to Series
series = cars.stack()
print(type(series))
print(series)
```

```python
#count----didnt count NAN values
print(series.count())
```

```python
#sum
print(series.sum())
```

```python
#product
print(series.product())
```

```python
#mean,median,mode,std,var

#mean
print(series.mean())

#median
print(series.median())

#mode
print(series.mode())

#std
print(series.std())

#var
print(series.var())
```

```python
#min and max

#min
print(series.min())

#max
print(series.max())
```

```python
#describe
print(series.describe())
```
