```python
student_dict = {
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students
```

```python
#rename----rename column name
students.rename(columns={'marks': 'percentage'}, inplace=True)
students
```

```python
#Maths Methods
#sum-- perform sum operation on all columns 
students.sum()

```

```python
#columnwise sum
students.sum(axis=0)

#rowwise sum--axis parameter
students.sum(axis=1)
```

```python
#mean--mean columnwise  
students.mean()

#mean--mean rowwise  
students.mean(axis=1)
```
